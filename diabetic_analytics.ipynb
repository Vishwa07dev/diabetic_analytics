```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("diabetese.tsv", sep='\t')
df.head()
```

```python
# Define the table columns 
df.info()
```

```python
# Let's find the mean age of all the patients
mean_age = df['AGE'].mean()
print(f"mean age is : {mean_age}")
```

```python
#Let's find the variance in the age of all the patients
# This helps me understand if all the patients are of similar age group or are they very different. It helps us understand the spread
var_age = df['AGE'].var()
print(f"Variance is : {var_age}")
```

```python
# Let's find the standard deviation in the age of all the patients
# SD also helps in finding the variance but in the same unit as mean. It tells us, how all elements are spread apart from mean
std_age = df['AGE'].std()
print(f"Standard deviation : {std_age}")
```

```python
print(f"For age -> mean : {mean_age} , variance : {var_age} , standard variation : {std_age}")
```

```python
### Understanding the boxplot

# Generate some example data
data = [75, 82, 88, 92, 96, 100, 105, 110, 120, 130, 135]

# Create a box plot
plt.boxplot(data)

# Add labels and a title
plt.xlabel('Data')
plt.ylabel('Values')
plt.title('Box Plot Example')

# Show the plot
plt.show()
```

```python
#Boxplot for the age field
plt.figure(figsize=(2,6))
plt.boxplot(df['AGE'], showmeans= True )
plt.grid(color='gray', linestyle = 'dotted')
plt.xlabel('Age')
plt.ylabel('years')
plt.show()
```

```python
#Boxplot of the BMI field
plt.figure(figsize=(6,2))
plt.boxplot(df['BMI'], vert=False, showmeans= True )
plt.grid(color='gray', linestyle = 'dotted')
plt.xlabel('Value')
plt.ylabel('BMI')
plt.show()
```

```python
#Boxplot of the BP field
plt.figure(figsize=(6,2))
plt.boxplot(df['BP'], vert=False, showmeans= True )
plt.grid(color='gray', linestyle = 'dotted')
plt.xlabel('Value')
plt.ylabel('BP')
plt.show()
```

```python
# Boxplot of the disease progression
plt.figure(figsize=(6,2))
plt.boxplot(df['Y'], vert=False, showmeans= True )
plt.grid(color='gray', linestyle = 'dotted')
plt.xlabel('Value')
plt.ylabel('Y')
plt.show()
```

```python
# I want to understand what age patients are mostly diabetic
# We can created histogram for the same
df['AGE'].hist(bins=12)
plt.title("Age distribution of the diabetic patients")
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python
# I want to understand what BMI value patients are mostly diabetic
# We can created histogram for the same
df['BMI'].hist(bins=12)
plt.title("BMI distribution of the diabetic patients")
plt.xlabel('BMI')
plt.ylabel('Count')
plt.show()
```

```python
# Scatter plot example
hours_studied = [2, 3, 1, 4, 2, 5, 4, 6, 3, 5]
exam_scores = [70, 80, 60, 85, 75, 90, 88, 92, 80, 95]

# Create a scatter plot
plt.scatter(hours_studied, exam_scores, color='blue', marker='o')

# Add labels and a title
plt.xlabel('Hours Studied')
plt.ylabel('Exam Scores')
plt.title('Scatter Plot: Hours Studied vs. Exam Scores')

# Show the plot
plt.show()
```
